## STEP 1 - Import Library
Mengimpor library yang dibutuhkan untuk proses data cleaning, yaitu pandas untuk manipulasi data, numpy untuk operasi numerik, dan scipy untuk fungsi statistik tambahan.

```python
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize
```

## STEP 2 - Load Dataset
Dataset housing_dirty.csv dimuat ke dalam DataFrame untuk diproses lebih lanjut.

```python
df = pd.read_csv('housing_dirty.csv')
df.head()
```

## STEP 3 - Eksplorasi Data
Melihat struktur dan kondisi awal dataset untuk mengetahui adanya missing values, tipe data, dan statistik dasar.

```python
df.info()
df.describe()
df.isnull().sum()
```

## STEP 4 - Hapus Duplikat
Menghapus data yang duplikat agar tidak terjadi redundansi dalam dataset.

```python
df.drop_duplicates(inplace=True)
print("Shape setelah hapus duplikat:", df.shape)
```

## STEP 5 - Normalisasi String
Membersihkan data string agar konsisten, seperti merapikan spasi dan menyamakan format huruf.

```python
df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()
```

## STEP 6 - Imputasi Missing Values
Mengisi nilai yang hilang menggunakan:
- Median untuk data numerik
- Modus untuk data kategorikal

```python
df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])
```

## STEP 7 - Menangani Outlier
Outlier ditangani menggunakan metode IQR untuk membatasi nilai yang terlalu ekstrem.

```python
for col in ['harga_juta', 'luas_m2']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)

# tambahan constraint
df['tahun_bangun'] = df['tahun_bangun'].clip(lower=1950)
```

## STEP 8 - Validasi Data
Memastikan bahwa data sudah bersih tanpa missing values dan duplikasi.

```python
print("Missing:", df.isnull().sum().sum())
print("Duplikat:", df.duplicated().sum())
print("Shape akhir:", df.shape)
```

## STEP 9 - Export Dataset
Menyimpan dataset yang sudah bersih ke dalam file baru.

```python
df.to_csv('housing_clean.csv', index=False)
```
